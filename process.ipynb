```python
import polars as pl
from tqdm.notebook import tqdm
from datetime import datetime
from dateutil.relativedelta import relativedelta
import os

def get_months_in_between(start_date: str, end_date: str):
    start = datetime.strptime(start_date.split('T')[0], "%Y-%m-%d")
    end = datetime.strptime(end_date.split('T')[0], "%Y-%m-%d")

    months = []
    while start <= end:
        months.append(start.strftime("%Y-%m"))
        start += relativedelta(months=1)
    return months

def expand_eventDates_chunked(df: pl.DataFrame, chunk_size=10000):
    num_rows = df.shape[0] 
    chunk_iter = df.iter_rows(named=True) 

    with tqdm(total=num_rows, desc="Expanding event dates", unit="rows") as pbar:
        while True:
            expanded_rows = []
            try:
                for _ in range(chunk_size):
                    row = next(chunk_iter)
                    occurrence_id = row["occurrenceID"]
                    verbatim_scientific_name = row["verbatimScientificName"]
                    latitude = row["decimalLatitude"]
                    longitude = row["decimalLongitude"]
                    uncertainty = row["coordinateUncertaintyInMeters"]
                    event_date = row["eventDate"]

                    if '/' in event_date:
                        start_date, end_date = event_date.split('/')
                        months = get_months_in_between(start_date, end_date)
                        for month in months:
                            expanded_rows.append({
                                'occurrenceID': occurrence_id,
                                'verbatimScientificName': verbatim_scientific_name,
                                'decimalLatitude': latitude,
                                'decimalLongitude': longitude,
                                'coordinateUncertaintyInMeters': uncertainty,
                                'eventDate': month
                            })
                    else:
                        event_date = datetime.strptime(start_date.split('T')[0], "%Y-%m-%d")
                        expanded_rows.append({
                            'occurrenceID': occurrence_id,
                            'verbatimScientificName': verbatim_scientific_name,
                            'decimalLatitude': latitude,
                            'decimalLongitude': longitude,
                            'coordinateUncertaintyInMeters': uncertainty,
                            'eventDate': event_date.strftime("%Y-%m")
                        })
                pbar.update(chunk_size)
            except StopIteration:
                if expanded_rows:
                    yield pl.DataFrame(expanded_rows)
                break

            yield pl.DataFrame(expanded_rows)

output_file = r'D:\Projects\Project-1\GBIF-Data\gbif-data-expanded.csv'

# if os.path.exists(output_file):
#     os.remove(output_file)
# first_chunk = True

# # Write the header for the first chunk
# for chunk in expand_eventDates_chunked(gbif_data, chunk_size=5000):
#     if first_chunk:
#         chunk.write_csv(output_file, include_header=True)  # Write the first chunk with headers
#         first_chunk = False
#     else:
#         # Append subsequent chunks without headers using a temporary file
#         temp_file = output_file + ".tmp"
#         chunk.write_csv(temp_file, include_header=False)  # Write to temp file
#         # Append temp file to the main output file
#         with open(temp_file, 'rb') as f_temp, open(output_file, 'ab') as f_output:
#             f_output.write(f_temp.read())  # Append the temp file content
#         os.remove(temp_file)  # Remove the temporary file after appending
```
